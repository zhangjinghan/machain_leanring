```python
import tensorflow as tf
#tf.__version__
from keras.models import Sequential
from keras.layers import Dense,Dropout
from keras.layers import Conv1D,GlobalAveragePooling1D,MaxPooling1D
import keras.metrics
from sklearn.model_selection import train_test_split
import numpy as np
import pandas as pd
import sklearn
from sklearn.preprocessing import normalize
import matplotlib.pyplot as plt
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score
from sklearn.metrics import recall_score
from sklearn.metrics import precision_score
from sklearn.metrics import f1_score
```

```python
#加载数据
train_data = pd.read_csv("hardware_train_set.csv")
test_data = pd.read_csv("hardware_test_set.csv")
#未处理的数据（有字符型数据）
train_data.head()
```

```python

train_label = train_data['label']
test_label = test_data['label']

print(train_label)

train_set = np.array(train_data).astype(np.float32)
train_set = train_set[:,1:]

test_set = np.array(train_data).astype(np.float32)
test_set = test_set[:,1:]

print(train_set)
```

```python
# 5 特征归一化  用MLP的时候用，LSTM不用
min_max_scaler = normalize.MinMaxScaler()
train_set = min_max_scaler.fit_transform(train_set)
test_set = min_max_scaler.fit_transform(test_set)
print("x_train_GA.shape",train_set.shape)


# 6 标签转化为one-hot编码 
y_train = keras.utils.to_categorical(train_label, num_classes=23)
y_test = keras.utils.to_categorical(test_label, num_classes=23)

print("y_train.shape",y_train.shape)
```

```python
cnn_1D=Sequential()
cnn_1D.add(Conv1D(64,1,activation='relu',input_shape=(108,1)))
cnn_1D.add(Conv1D(64,1,activation='relu'))
cnn_1D.add(MaxPooling1D(3))
cnn_1D.add(Conv1D(64,1,activation='relu'))
cnn_1D.add(Conv1D(64,1,activation='relu'))
cnn_1D.add(GlobalAveragePooling1D())
cnn_1D.add(Dropout(0.5))
cnn_1D.add(Dense(11,activation='softmax'))
cnn_1D.compile(
    loss='sparse_categorical_crossentropy',
    optimizer='rmsprop',
    metrics=(['accuracy'])
)
```
